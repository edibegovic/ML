```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

First we generate some training, validation and test data:

```python
N = 20
X = np.linspace(0, 1, N)

np.random.seed(10)

t_train = np.sin(2*np.pi*X) + np.random.normal(scale=0.25, size=N)
t_test = np.sin(2*np.pi*X) + np.random.normal(scale=0.25, size=N)
t_val = np.sin(2*np.pi*X) + np.random.normal(scale=0.25, size=N)


# Make the data column vectors
X = X[:, np.newaxis]
t_train = t_train[:, np.newaxis]
t_test = t_test[:, np.newaxis]
t_val = t_val[:, np.newaxis]
```

```python
# Input and output
x = tf.placeholder(tf.float32, [None, 1])
t = tf.placeholder(tf.float32, [None, 1])

# Defined the model parameters
W1 = tf.get_variable("W1", [1, 1000], initializer=tf.random_normal_initializer)
b1 = tf.get_variable("b1", [1000], initializer=tf.random_normal_initializer)
W2 = tf.get_variable("W2", [1000, 1], initializer=tf.random_normal_initializer)
b2 = tf.get_variable("b2", [1], initializer=tf.random_normal_initializer)
    
# Construct model
z1 = tf.nn.relu(tf.matmul(x, W1) + b1)
y = tf.matmul(z1, W2) + b2

# Define the loss function
loss = tf.losses.mean_squared_error(labels=t, predictions=y)

# Define the optimizer operation
learning_rate = tf.placeholder(tf.float32)
optimizer = tf.train.AdamOptimizer(learning_rate = learning_rate).minimize(loss)

# Define the optimizer operation
learning_rate = tf.placeholder(tf.float32)
optimizer = tf.train.AdamOptimizer(learning_rate = learning_rate).minimize(loss)

# Make an operation that initializes the variables
init = tf.global_variables_initializer()
```

```python
X_test = np.linspace(0, 1, 100)[:, np.newaxis]
train_loss_list = []
val_loss_list = []

best_val_loss = np.float('inf')
saver = tf.train.Saver()

n_epoch = 10000

# Start a new session
with tf.Session() as session:
    # Initialize the values
    session.run(init)
    
    # Training cycle
    for epoch in range(n_epoch):
        _, loss_value = session.run([optimizer, loss], feed_dict={x: X, t: t_train, learning_rate: 0.0001})
        val_loss_value = session.run(loss, feed_dict={x: X, t: t_val})
        
        if val_loss_value < best_val_loss:
            best_val_loss = val_loss_value
            saver.save(sess=session, save_path="./model")
        
        train_loss_list.append(loss_value)
        val_loss_list.append(val_loss_value)
        
        if epoch % 1000 == 0:
            print("Epoch: ", epoch, "loss =", loss_value)            
            
    print("Optimization done")

    # Make a prediction at the test points
    y_test_final = session.run(y, feed_dict={x: X_test})
```

Then we plot the losses.

```python
fig, ax = plt.subplots(figsize=(15, 6))
ax.plot(range(n_epoch), train_loss_list, label="Training loss")
ax.plot(range(n_epoch), val_loss_list, label="Validation loss")
ax.legend()
ax.set_ylim(0, 0.2)
ax.set_xlabel(r'epoch')
ax.set_ylabel(r'loss')
plt.show()
```

Now lest plot the fitted function

```python
with tf.Session() as session:
    saver.restore(session, "./model")
    y_test_optimal = session.run(y, feed_dict={x: X_test})


fig, ax = plt.subplots(1,2, figsize=(15, 6))
ax[0].scatter(X, t_train, color='C0', label="Training data")
ax[0].plot(X_test, y_test_final, color='C2', label="Final learned function")
ax[0].plot(X_test, y_test_optimal, color='C4', label="Minimal val loss function")
ax[0].legend()
ax[0].set_xlabel(r'$x$')
ax[0].set_ylabel(r'$t$')
ax[0].set_title(r'Training')

ax[1].scatter(X, t_val, color='C1', label="Val data")
ax[1].plot(X_test, y_test_final, color='C2', label="Final learned function")
ax[1].plot(X_test, y_test_optimal, color='C4', label="Minimal val loss function")
ax[1].legend()
ax[1].set_xlabel(r'$x$')
ax[1].set_ylabel(r'$t$')
ax[1].set_title(r'Val')
plt.show()
```
