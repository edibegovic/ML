```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

First we generate some training and test data:

```python
N = 20
X = np.linspace(0, 1, N)

np.random.seed(10)

t_train = np.sin(2*np.pi*X) + np.random.normal(scale=0.3, size=N)
t_test = np.sin(2*np.pi*X) + np.random.normal(scale=0.3, size=N)

# Make the data column vectors
X = X[:, np.newaxis]
t_train = t_train[:, np.newaxis]
t_test = t_test[:, np.newaxis]
```

```python
fig, ax = plt.subplots()
ax.scatter(X, t_train, label="Training data")
ax.scatter(X, t_test, label="Test data")
ax.legend()
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$t$')
plt.show()
```

```python
# Input and output
x = tf.placeholder(tf.float32, [None, 1])
t = tf.placeholder(tf.float32, [None, 1])

# Defined the model parameters
W1 = tf.get_variable("W1", [1, 500], initializer=tf.random_normal_initializer)
b1 = tf.get_variable("b1", [500], initializer=tf.random_normal_initializer)
W2 = tf.get_variable("W2", [500, 1], initializer=tf.random_normal_initializer)
b2 = tf.get_variable("b2", [1], initializer=tf.random_normal_initializer)
    
# Construct model
z1 = tf.nn.relu(tf.matmul(x, W1) + b1)
y = tf.matmul(z1, W2) + b2

# Define the loss function
reg_fact = tf.placeholder(tf.float32)
mse = tf.losses.mean_squared_error(labels=t, predictions=y)
loss = mse + reg_fact*(tf.nn.l2_loss(W1) + tf.nn.l2_loss(W2))

# Define the optimizer operation
learning_rate = tf.placeholder(tf.float32)
optimizer = tf.train.AdamOptimizer(learning_rate = learning_rate).minimize(loss)

# Make an operation that initializes the variables
init = tf.global_variables_initializer()
```

```python
X_test = np.linspace(0, 1, 100)[:, np.newaxis]

# Start a new session
with tf.Session() as session:
    # Initialize the values
    session.run(init)
    
    # Training cycle
    for epoch in range(20000):
        _, loss_value = session.run([optimizer, loss], feed_dict={x: X, t: t_train, learning_rate: 0.001, reg_fact: 0.002})
        
        if epoch % 1000 == 0:
            print("Epoch: ", epoch, "loss =", loss_value)            
            
    print("Optimization done")

    # Make a prediction at the test points
    y_test = session.run(y, feed_dict={x: X_test})
```

```python
X_ = np.linspace(0, 1, 100)
y_ = np.sin(2*np.pi*X_)

fig, ax = plt.subplots(1,2, figsize=(15, 6))
ax[0].scatter(X, t_train, color='C0', label="Training data")
ax[0].plot(X_test, y_test, color='C2', label="Learned function")
ax[0].plot(X_, y_, color='C5', label="sin")
ax[0].legend()
ax[0].set_xlabel(r'$x$')
ax[0].set_ylabel(r'$t$')
ax[0].set_title(r'Training')

ax[1].scatter(X, t_test, color='C1', label="Test data")
ax[1].plot(X_test, y_test, color='C2', label="Learned function")
ax[1].plot(X_, y_, color='C5', label="sin")
ax[1].legend()
ax[1].set_xlabel(r'$x$')
ax[1].set_ylabel(r'$t$')
ax[1].set_title(r'Test')
plt.show()
```

Now let us see if we can find a optimal $\lambda$ value.

First we make a validation set.

```python
t_val = np.sin(2*np.pi*X) + np.random.normal(scale=0.2, size=N)[:, np.newaxis]
```

```python
reg_fact_values = np.logspace(-5, -1, 6)
train_mse_list = []
val_mse_list = []

for reg_fact_value in reg_fact_values:
    print("lambda =", reg_fact_value)
    with tf.Session() as session:
        session.run(init)
        for epoch in range(30000):
            _, loss_value = session.run([optimizer, loss], feed_dict={x: X, t: t_train, learning_rate: 0.001, reg_fact: reg_fact_value})
            if epoch % 2000 == 0:
                print("Epoch: ", epoch, "loss =", loss_value)
                
        # Make a prediction at the val points
        train_mse = session.run(mse, feed_dict={x: X, t: t_train})
        val_mse = session.run(mse, feed_dict={x: X, t: t_val})
        train_mse_list.append(train_mse)
        val_mse_list.append(val_mse)
```

```python
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(reg_fact_values, train_mse_list, label="Training MSE")
ax.plot(reg_fact_values, val_mse_list, label="Val MSE")
ax.legend()
ax.set_xlabel(r'$\lambda$')
ax.set_ylabel(r'$MSE$')
ax.set_title(r'Training')
ax.set_xscale("log")
plt.show()
```
